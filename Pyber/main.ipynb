```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
merged_data = pd.merge(ride_data, city_data, on="city")

# Display the data table for preview
merged_data.head()
```

## Bubble Plot of Ride Sharing Data

```python
#Creates dataframes for future use in the rest of the code
urb = merged_data.loc[merged_data["type"] == "Urban"]
rural = merged_data.loc[merged_data["type"] == "Rural"]
subUrb = merged_data.loc[merged_data["type"] == "Suburban"]

avgFare = round(merged_data.groupby(['city'])['fare'].mean(), 2)
avgUrb = round(urb.groupby(['city'])['fare'].mean(), 2)
avgRur = round(rural.groupby(['city'])['fare'].mean(), 2)
avgSub = round(subUrb.groupby(['city'])['fare'].mean(), 2)

totalRides = merged_data.groupby(['city'])['ride_id'].count()
urbRides = urb.groupby(['city'])['ride_id'].count()
rurRides = rural.groupby(['city'])['ride_id'].count()
subRides = subUrb.groupby(['city'])['ride_id'].count()

totalDrivers = merged_data.groupby(['city'])['driver_count'].mean()
urbDrivers = urb.groupby(['city'])['driver_count'].mean()
rurDrivers = rural.groupby(['city'])['driver_count'].mean()
subDrivers = subUrb.groupby(['city'])['driver_count'].mean()
```

```python
# Obtain the x and y coordinates for each of the three city types
urbX = urbRides
urbY = avgUrb
rurX = rurRides
rurY = avgRur
subX = subRides
subY = avgSub

# Build the scatter plots for each city types
plt.title('Average Fares By City Type');
plt.scatter(urbX, urbY, s = 5*totalDrivers,c="gold",marker="o",edgecolors="black", label="Urban")
plt.scatter(rurX, rurY, s = 5*totalDrivers,c="skyblue",marker="o",edgecolors="black", label="Suburban")
plt.scatter(subX, subY, s = 5*totalDrivers,c="coral",marker="o",edgecolors="black", label="Rural")

# Incorporate the other graph properties
plt.grid()
plt.xlabel("Number of Rides")
plt.ylabel("Average Price")

# Create a legend
plt.legend(loc="best")

# Incorporate a text label regarding circle size
plt.text(42, 42, "Note: Circle sizes based on amount of drivers per city")


# Save Figure
plt.savefig("../Images/bubbleChart.png")
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
urbanFares = urb['fare'].sum()
suburbanFares = subUrb['fare'].sum()
ruralFares = rural['fare'].sum()

# Build Pie Chart
fares = [urbanFares, suburbanFares, ruralFares]
labels = ["Urban", "Suburban", "Rural"]
colors = ["gold", "skyblue", "coral"]
explode = (0.1, 0, 0)

plt.pie(fares, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=280)

plt.axis("equal")
plt.title("% of Total Fares by City Type")

# Save Figure
plt.savefig("../Images/fareChart.png")

# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
urbRides = urb["ride_id"].count()
subRides = subUrb["ride_id"].count()
rurRides = rural["ride_id"].count()

# Build Pie Chart
rides = [urbRides, subRides, rurRides]
labels = ["Urban", "Suburban", "Rural"]
colors = ["gold", "skyblue", "coral"]
explode = (0, 0.1, 0)

plt.pie(rides, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=280)

plt.axis("equal")
plt.title("% of Total Rides by City Type")

# Save Figure
plt.savefig("../Images/rideChart.png")

# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
urbDrivers = urb["driver_count"].sum()
subDrivers = subUrb["driver_count"].sum()
rurDrivers = rural["driver_count"].sum()

# Build Pie Chart
drivers = [urbDrivers, subDrivers, rurDrivers]
labels = ["Urban", "Suburban", "Rural"]
colors = ["gold", "skyblue", "coral"]
explode = (0, 0, 0.1)

plt.pie(drivers, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=280)

plt.axis("equal")
plt.title("% of Total Drivers by City Type")

# Save Figure
plt.savefig("../Images/driverChart.png")

# Show Figure
plt.show()
```
